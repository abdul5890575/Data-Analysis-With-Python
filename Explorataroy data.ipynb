### ANOVA

```python
df_test = df[['drive-wheels','body-style','price']]
test2=df_test[['drive-wheels', 'price']].groupby(['drive-wheels'])
test2.head(2)
```

```python
test2.get_group('4wd')['price']
```

```python
f,p=stats.f_oneway(test2.get_group('fwd')['price'],test2.get_group('rwd')['price'])
print(f,p)
```

```python
f,p=stats.f_oneway(test2.get_group('fwd')['price'],test2.get_group('4wd')['price'])
print(f,p)
```

```python
f,p=stats.f_oneway(test2.get_group('4wd')['price'],test2.get_group('rwd')['price'])
print(f,p)
```

```python
f,p=stats.f_oneway(test2.get_group('4wd')['price'],test2.get_group('fwd')['price'])
print(f,p)
```
